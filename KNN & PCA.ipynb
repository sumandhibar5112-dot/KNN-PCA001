Question 1: What is K-Nearest Neighbors (KNN) and how does it work in both
classification and regression problems?

**K-Nearest Neighbors (KNN)** is a simple, non-parametric, instance-based machine learning algorithm used for **classification and regression**. It makes predictions based on the idea that *similar data points exist close to each other in feature space*.

---

#  **How KNN Works (General Idea)**

1. **Choose the value of K**
   → K = number of neighbors to consider (e.g., K=3 or K=5).

2. **Calculate distance** between the new data point and all existing training points.
   Common choices:

   * Euclidean distance (most common)
   * Manhattan distance
   * Minkowski distance

3. **Find the K closest points** to the new data point.

4. **Make a prediction** based on these K neighbors.

---

#  **KNN for Classification**

KNN assigns the class **based on majority voting** among the K nearest neighbors.

###  Example:

Suppose K=5 and the classes of the nearest neighbors are:

* Class A: 3 points
* Class B: 2 points

Prediction → **Class A** (majority vote)

### Characteristics in classification:

* Decision boundaries are often irregular and nonlinear.
* Sensitive to class imbalance.
* Works well when classes are well-separated.

---

#  **KNN for Regression**

For regression tasks, KNN predicts a **numerical value** by averaging the values of the K nearest neighbors.

###  Example:

Values of 5 nearest neighbors:
5.2, 4.8, 5.0, 5.5, 4.9

Prediction → **Mean = 5.08**

### Other averaging options:

* Weighted average (closer neighbors get higher weight)
* Median of the neighbors

---

#  Advantages of KNN

* Simple and easy to understand.
* No training time → all computation happens at prediction time.
* Works well for non-linear problems.

---

#  Limitations

* Slow for large datasets (distance calculated for every prediction).
* Performance drops in high-dimensional data (curse of dimensionality).
* Sensitive to noisy features and irrelevant features.
* Requires proper scaling/normalization.

---



Question 2: What is the Curse of Dimensionality and how does it affect KNN
performance?

The **Curse of Dimensionality** refers to a set of problems that arise when data has **too many features (dimensions)**. As dimensionality increases, data becomes sparse, distance metrics lose meaning, and algorithms that rely on distance—like **KNN**—perform poorly.

---

#  **What Is the Curse of Dimensionality?**

When the number of features grows:

### 1. **Data becomes sparse**

* Points are spread out in high-dimensional space.
* Harder to find “close” neighbors because everything becomes far apart.

### 2. **Distance metrics lose reliability**

* In high dimensions, the distance between the closest and farthest points becomes similar.
* Meaning:
  **"Near" and "far" lose meaning → KNN cannot identify true neighbors.**

### 3. **Exponential increase in volume**

* To maintain the same density, the number of samples needed grows exponentially with dimensions.

---

#  **How Curse of Dimensionality Affects KNN**

KNN relies entirely on **distance** to find neighbors.
When dimensionality is high:

---

## **1. All points become equally distant**

* Euclidean distance becomes almost uniform.
* KNN cannot differentiate close points from far ones → bad predictions.

---

## **2. Overfitting increases**

* KNN becomes too sensitive to noise and irrelevant features.
* A single noisy feature can dominate the distance calculation.

---

## **3. Requires exponentially more data**

* High-dimensional spaces need far more samples for meaningful patterns.
* With insufficient data, KNN cannot generalize well.

---

## **4. Computation becomes expensive**

* Distance calculation requires looping through all features.
* Higher dimensions → slower predictions.

---

#  **Example**

In 2D, you may need ~100 points for good density.
In 100D, you might need **millions** to achieve the same density.

This is why high-dimensional data (like text or genetics) requires dimensionality reduction before KNN.

---

#  **How to Reduce the Curse of Dimensionality in KNN**

* **Feature selection** (remove irrelevant features)
* **Dimensionality reduction**

  * PCA (Principal Component Analysis)
  * t-SNE / UMAP
* **Normalize** or **standardize** features
* **Use distance-weighted KNN**

---




Question 3: What is Principal Component Analysis (PCA)? How is it different from
feature selection?

**Principal Component Analysis (PCA)** is a **dimensionality reduction** technique that transforms high-dimensional data into a smaller set of new features called **principal components**, while preserving as much variance (information) as possible.

---

#  **What is PCA?**

PCA is a **linear transformation** method that:

1. Identifies the directions (components) in the data where the **variance is maximum**.
2. Creates new features (principal components) that are:

   * **Uncorrelated**
   * Ordered by how much variance they capture
3. Reduces dimensions by keeping only the top components that explain most of the variance.

### **Example**

If you have a dataset with 50 features, PCA might find that only 5 components explain 95% of the variance → you can reduce the data from 50 → 5 dimensions.

---

#  **How PCA Works (Conceptually)**

1. **Standardize** the data
2. Compute **covariance matrix**
3. Find **eigenvalues and eigenvectors**
4. Sort eigenvectors by largest eigenvalues
5. Choose top k eigenvectors → new feature space

Principal components are **linear combinations** of original features.

---

#  **How PCA Differs from Feature Selection**

| Aspect                      | **PCA (Dimensionality Reduction)**                | **Feature Selection**                                     |
| --------------------------- | ------------------------------------------------- | --------------------------------------------------------- |
| **What it does**            | Creates *new* features (principal components)     | Selects a subset of *existing* features                   |
| **Keeps original meaning?** | No — transformed features lose physical meaning | ✔ Yes — original features remain                          |
| **Method type**             | Feature extraction                                | Feature selection                                         |
| **How it reduces features** | Combines correlated features into components      | Removes irrelevant or redundant features                  |
| **Data transformation**     | ✔ Yes, transforms the feature space               |  No transformation                                       |
| **Interpretability**        | Low (components are combinations of features)     | High (keeps human-understandable features)                |
| **Goal**                    | Capture maximum variance in fewer dimensions      | Improve accuracy, reduce noise, keep important predictors |

---

#  **In Short**

### **PCA**

➡ Reduces dimensionality by **creating new features** (principal components) that maximize variance.
➡ Useful when features are highly correlated.

### **Feature Selection**

➡ Reduces dimensionality by **choosing the best existing features**.
➡ Keeps data interpretable and original.

---




Question 4: What are eigenvalues and eigenvectors in PCA, and why are they
important?

In **Principal Component Analysis (PCA)**, **eigenvalues** and **eigenvectors** are the mathematical foundation that determine the **directions** of maximum variance in the data and **how much variance** each direction captures.

---

#  **What Are Eigenvectors and Eigenvalues?**

### **Eigenvectors**

* Directions in the feature space along which the data varies the most.
* In PCA, each eigenvector corresponds to a **principal component**.
* They represent **new axes** after transformation.

### **Eigenvalues**

* Measure how much **variance** (information) is captured by each eigenvector.
* A large eigenvalue → the corresponding eigenvector captures more variance.
* Helps determine how many components to keep.

---

#  **Why Are They Important in PCA?**

### **1. They identify the directions of maximum variance**

PCA looks for directions where the data spreads out the most.
These directions are found by computing the **eigenvectors** of the covariance matrix.

### **2. Eigenvalues tell us which components are important**

A principal component with a higher eigenvalue:

* Explains more variance
* Contains more information
* Should be kept in dimensionality reduction

Components with very small eigenvalues add little information and can be removed.

### **Example:**

If the eigenvalues are:

* λ₁ = 5.2
* λ₂ = 1.1
* λ₃ = 0.02

Then most of the variance is in **component 1**, then **component 2**, and component 3 can be ignored.

---

#  **PCA Workflow with Eigenvalues/Eigenvectors**

1. Compute covariance matrix of the data.
2. Find **eigenvalues** and **eigenvectors** of this matrix.
3. Sort eigenvalues (and corresponding eigenvectors) in descending order.
4. Top-k eigenvectors become the **principal components**.
5. Project data onto these components to get reduced dimensions.

---

#  **Intuitive Explanation**

Imagine you have a cloud of points:

* The **eigenvectors** show the directions in which the cloud is stretched.
* The **eigenvalues** tell you *how much* stretching occurs in each direction.

PCA keeps the directions with the most stretching (information) and drops the rest.

---

#  **In Short**

| Concept         | Meaning                       | Role in PCA                         |
| --------------- | ----------------------------- | ----------------------------------- |
| **Eigenvector** | Direction of maximum variance | Defines principal component         |
| **Eigenvalue**  | Amount of variance captured   | Helps rank components (PC1, PC2, …) |

---



Question 5: How do KNN and PCA complement each other when applied in a single
pipeline?

KNN and PCA often work **extremely well together** because PCA solves many of the problems that hurt KNN. Using them in the same pipeline improves **accuracy, speed, and robustness**—especially with high-dimensional data.

---

#  **How KNN and PCA Complement Each Other**

## **1. PCA reduces dimensionality → KNN works better**

KNN struggles in high-dimensional spaces due to the **curse of dimensionality**:

* Distances become less meaningful
* Data becomes sparse
* KNN becomes slow and inaccurate

**PCA solves this** by projecting data into a lower-dimensional space while preserving most variance.

### Result:

 Distances become more meaningful
 KNN classification/regression becomes more accurate

---

## **2. PCA removes noise and redundancy → cleaner input for KNN**

Many original features may be:

* Correlated
* Noisy
* Irrelevant

PCA transforms them into **uncorrelated principal components**.

### Result:

 KNN focuses on important patterns
Less noise → fewer wrong neighbors
 Improved generalization

---

## **3. PCA improves KNN speed**

KNN is slow at prediction time because it must compute distances to *every training point* across *all features*.

PCA reduces the number of features from, say, 300 → 20.

### Result:

 Much faster distance computation
 Faster predictions
 More scalable

---

## **4. PCA can mitigate overfitting in KNN**

KNN is non-parametric and very sensitive to irrelevant features.

PCA keeps components with high variance and drops those with tiny eigenvalues (low info).

### Result:

 Lower overfitting
 More stable KNN decisions

---

#  **PCA + KNN Pipeline Example**

1. **Standardize the data**
2. **Apply PCA** (e.g., keep components explaining 95% variance)
3. **Train KNN** on the transformed data
4. **Use the same PCA model to transform test data**
5. **Make predictions with KNN**

This is often implemented using:

```python
Pipeline([("scaler", StandardScaler()),
          ("pca", PCA(n_components=0.95)),
          ("knn", KNeighborsClassifier(n_neighbors=5))])
```

---

#  **Summary: Why They Work Well Together**

| Problem in High Dimensions | How PCA Helps                   | KNN Benefit                 |
| -------------------------- | ------------------------------- | --------------------------- |
| Distance loses meaning     | Reduces dimensions              | Neighbors become meaningful |
| Redundant/noisy features   | Removes correlation & noise     | Better accuracy             |
| Slow distance calculation  | Fewer features                  | Faster prediction           |
| Overfitting risk           | Keeps only informative variance | More robust model           |

---

#  Final Takeaway

**PCA enhances KNN by reducing dimensionality, eliminating noise, and making distances meaningful.**
Together, they form a powerful pipeline for high-dimensional datasets like images, text embeddings, and gene data.


````python
#Question 6: Train a KNN Classifier on the Wine dataset with and without feature scaling. Compare model accuracy in both cases.
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# 1. Load and Prepare Data
wine = load_wine()
X, y = wine.data, wine.target

# Split the data into training and testing sets (using a fixed random_state for reproducibility)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- A Visual Guide to KNN and Scaling ---
#

# 2. Unscaled KNN
print("--- Unscaled KNN Classification ---")
# Initialize the KNN model (using k=5, a common default)
knn_unscaled = KNeighborsClassifier(n_neighbors=5)

# Train the model on the raw, unscaled training data
knn_unscaled.fit(X_train, y_train)

# Make predictions
y_pred_unscaled = knn_unscaled.predict(X_test)

# Calculate accuracy
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)
print(f"Accuracy (Unscaled Data): {accuracy_unscaled:.4f}")
print("-" * 40)

# 3. Scaled KNN
print("--- Scaled KNN Classification ---")
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler ONLY on the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize a new KNN model
knn_scaled = KNeighborsClassifier(n_neighbors=5)

# Train the model on the SCALED training data
knn_scaled.fit(X_train_scaled, y_train)

# Make predictions
y_pred_scaled = knn_scaled.predict(X_test_scaled)

# Calculate accuracy
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print(f"Accuracy (Scaled Data): {accuracy_scaled:.4f}")
print("-" * 40)

# 4. Compare Results
print("--- Comparison ---")
print(f"Unscaled Accuracy: {accuracy_unscaled:.4f}")
print(f"Scaled Accuracy:   {accuracy_scaled:.4f}")

if accuracy_scaled > accuracy_unscaled:
    print("\nConclusion: Feature scaling significantly improved the model accuracy.")
elif accuracy_scaled < accuracy_unscaled:
    print("\nConclusion: Feature scaling slightly decreased the model accuracy (though scaling is generally recommended for KNN).")
else:
    print("\nConclusion: Feature scaling had no significant effect on the model accuracy.")
````

````python
#Question 7: Train a PCA model on the Wine dataset and print the explained variance ratio of each principal component.
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and Scale Data
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

# Scaling is crucial for PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Train PCA Model
# Initialize PCA to retain all components (13 components for the Wine dataset)
#
pca = PCA(n_components=None) # n_components=None means all components are kept

# Fit PCA on the scaled data
pca.fit(X_scaled)

# 3. Analyze Explained Variance
print("--- PCA Explained Variance Analysis (Wine Dataset) ---")

# Explained Variance Ratio for each Principal Component
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio of Each Principal Component:")

# Create a clear table or formatted output
explained_df = pd.DataFrame({
    'Component': [f'PC{i+1}' for i in range(len(explained_variance))],
    'Explained Variance Ratio': explained_variance,
    'Cumulative Ratio': np.cumsum(explained_variance)
})

# Format the floats for readability
explained_df['Explained Variance Ratio'] = explained_df['Explained Variance Ratio'].map('{:.4f}'.format)
explained_df['Cumulative Ratio'] = explained_df['Cumulative Ratio'].map('{:.4f}'.format)

print(explained_df)
print("\n" + "="*60)

# Key Interpretations
# The first component (PC1) captures the most variance
pc1_variance = pca.explained_variance_ratio_[0]
print(f"The First Principal Component (PC1) explains {pc1_variance*100:.2f}% of the total variance.")

# Determine how many components are needed to explain a high percentage (e.g., 90%)
cumulative_variance = np.cumsum(explained_variance)
n_components_90 = np.argmax(cumulative_variance >= 0.90) + 1 # +1 for 1-based indexing

print(f"The first {n_components_90} components capture over 90% of the total variance.")
print("This suggests that the 13 original features can be effectively reduced to this smaller number.")
print("="*60)

# Optional: Plot the explained variance (Scree Plot)
plt.figure(figsize=(8, 5))
plt.plot(np.arange(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--', color='blue', label='Individual Variance')
plt.plot(np.arange(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='-', color='red', label='Cumulative Variance')
plt.title('Scree Plot: Explained Variance by Principal Component')
plt.xlabel('Principal Component Number')
plt.ylabel('Explained Variance Ratio')
plt.xticks(np.arange(1, len(explained_variance) + 1, 1))
plt.grid(True)
plt.legend()
plt.show()
````

````python
#Question 8: Train a KNN Classifier on the PCA-transformed dataset (retain top 2 components). Compare the accuracy with the original dataset.
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

# --- 1. Data Preparation (Load, Scale, and Split) ---
wine = load_wine()
X, y = wine.data, wine.target

# Split data first
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale the data (Crucial for both KNN and PCA)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- 1. Data Prepared ---")

# --- 2. Benchmark KNN (Full Scaled Features - 13 dimensions) ---
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

print(f"Benchmark Accuracy (KNN on 13 Scaled Features): {accuracy_scaled:.4f}")
print("-" * 60)

# --- 3. PCA Transformation (Retain top 2 components) ---
# Initialize PCA to retain only the top 2 components
pca = PCA(n_components=2)

# Fit PCA on the scaled training data and transform both train and test sets
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Check the variance explained by the 2 components
explained_variance_ratio_2 = pca.explained_variance_ratio_.sum()

print("--- 3. PCA Transformation Complete ---")
print(f"Total variance explained by 2 components (PC1 + PC2): {explained_variance_ratio_2:.4f}")
print(f"New feature matrix shape (Training): {X_train_pca.shape}")
print("-" * 60)

# --- 4. PCA Model KNN (2 Dimensions) ---
knn_pca = KNeighborsClassifier(n_neighbors=5)
# Train the model on the transformed data (2 principal components)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)

print(f"PCA Accuracy (KNN on 2 PCA Features): {accuracy_pca:.4f}")
print("-" * 60)

# --- 5. Comparison ---
print("--- Comparison of Model Performance ---")
print(f"Accuracy with 13 Scaled Features: {accuracy_scaled:.4f}")
print(f"Accuracy with 2 PCA Features:     {accuracy_pca:.4f}")

# Calculate the difference
accuracy_difference = accuracy_scaled - accuracy_pca

print(f"\nDifference (Scaled - PCA): {accuracy_difference:.4f}")

if accuracy_difference < 0.05:
    print("\nConclusion: The model using only 2 principal components achieved near-identical accuracy to the model using all 13 features!")
else:
     print("\nConclusion: While the model using 2 principal components performed well, there was a larger drop in accuracy compared to the full feature set.")
````

````python
#Question 9: Train a KNN Classifier with different distance metrics (euclidean, manhattan) on the scaled Wine dataset and compare the results.
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# 1. Data Preparation (Load, Scale, and Split)
wine = load_wine()
X, y = wine.data, wine.target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale the data (Essential for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- Data Prepared (Scaled) ---")

# --- 2. Euclidean KNN (Metric: 'minkowski' with p=2) ---
# Euclidean distance is the straight-line distance, defined by the formula:
# $d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}$
#
knn_euclidean = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)

# Train the model
knn_euclidean.fit(X_train_scaled, y_train)

# Make predictions and calculate accuracy
y_pred_euclidean = knn_euclidean.predict(X_test_scaled)
accuracy_euclidean = accuracy_score(y_test, y_pred_euclidean)

print(f"Accuracy (Euclidean Distance): {accuracy_euclidean:.4f}")
print("-" * 50)

# --- 3. Manhattan KNN (Metric: 'minkowski' with p=1) ---
# Manhattan distance is the distance calculated along the grid-like path (sum of absolute differences), defined by:
# $d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n} |x_i - y_i|$
knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric='manhattan')

# Train the model
knn_manhattan.fit(X_train_scaled, y_train)

# Make predictions and calculate accuracy
y_pred_manhattan = knn_manhattan.predict(X_test_scaled)
accuracy_manhattan = accuracy_score(y_test, y_pred_manhattan)

print(f"Accuracy (Manhattan Distance): {accuracy_manhattan:.4f}")
print("-" * 50)

# --- 4. Comparison ---
print("--- Accuracy Comparison ---")
print(f"Euclidean (L2) Accuracy: {accuracy_euclidean:.4f}")
print(f"Manhattan (L1) Accuracy: {accuracy_manhattan:.4f}")

if accuracy_euclidean > accuracy_manhattan:
    print("\nObservation: Euclidean distance provided slightly higher accuracy for this dataset.")
elif accuracy_manhattan > accuracy_euclidean:
    print("\nObservation: Manhattan distance provided slightly higher accuracy for this dataset.")
else:
    print("\nObservation: Both distance metrics yielded the exact same accuracy for this dataset.")
````

````python
#Question 10: You are working with a high-dimensional gene expression dataset to classify patients with different types of cancer. Due to the large number of features and a small number of samples, traditional models overfit.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# --- 1. Simulation of High-Dimensional Gene Expression Data ---
# 100 samples (patients), 500 features (genes), but only 20 are actually informative
# This mimics the "Curse of Dimensionality"
X, y = make_classification(
    n_samples=100,
    n_features=500,
    n_informative=20,
    n_redundant=30,
    n_classes=2,
    random_state=42
)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Original Feature Shape: {X_train.shape} (High Dimensionality)")

# --- 2. Scaling (Essential for PCA) ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- 3. PCA for Dimensionality Reduction ---
# Strategy: Keep enough components to explain 90% of the variance
pca = PCA(n_components=0.90)

# Fit PCA on training data and transform both sets
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

n_components_kept = pca.n_components_
print(f"Reduced Feature Shape:  {X_train_pca.shape}")
print(f"Components retained to explain 90% variance: {n_components_kept}")
print("-" * 50)

#

# --- 4. KNN Classification on Reduced Data ---
# Using KNN on the Principal Components
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_pca, y_train)
y_pred = knn.predict(X_test_pca)

# --- 5. Evaluation ---
print("--- Model Evaluation (PCA + KNN) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# --- 6. Justification Logic (for output) ---
print("-" * 50)
print("--- Justification for Stakeholders ---")
print(f"1. Compression: We reduced {X_train.shape[1]} noisy gene features down to just {n_components_kept} components.")
print(f"2. Signal Retention: These {n_components_kept} components still capture 90% of the biological signal.")
print("3. Robustness: By removing noise, KNN can now identify patient similarities more reliably.")
````
